```python
# Importing the necessary libraries required
# Importing the pandas library
#
import pandas as pd

# Importing the numpy library
#
import numpy as np
```

```python
# Loading the CSV file into a dataframe
# Storing the url location of the dataset to the variable url
url = 'http://bit.ly/autolib_dataset'

# read the dataset from the url above and store the data frame in the variable df
df = pd.read_csv(url,encoding="latin1")

# Preview the first 10 rows
df.head(10)
```

```python
# Getting more information about the dataset df.
df.info()
# The dataset has 5000 rows and 25 columns
```

```python
# Check the shape of the data provided.
df.shape
# The dataset has 5000 rows and 25 columns
```

```python
# Describing the data more
df.describe()
```

```python
# Data Validity : Deleting irrelevant Data
# Drop/delete the "Cars" column since it's redundant to the "Bluecar counter" column.
cardrop_df = df.drop(['Cars'], axis = 1)
cardrop_df.head()
```

```python
# Check for any missing data
cardrop_df.isnull().sum()
```

```python
# Data Completeness : Dropping columns with missing values
# Drop/delete the "Displayed comment" column because it has a lot of missing values.
# Drop/delete the "Scheduled at" column because it has a lot of missing values.
missing_df = cardrop_df.drop(['Displayed comment','Scheduled at'], axis = 1)
missing_df.head()
```

```python
# Data Consistency - any repeated data points
# checking for duplicates in the dataframe
missing_df.duplicated().sum()

# There are no duplicate values found
```

```python
# Data Uniformity : Renaming the columns
renamed_df = missing_df.rename(columns = {'Bluecar counter':'Bluecars','Utilib counter':'Utilib','Utilib 1.4 counter':'Utilibtwo', 'Charge Slots':'ChargeSlots','Charging Status':'ChargingStatus','Geo point':'Geo','Postal code':'Postal','Public name':'PublicName','Rental status':'RentalStatus','Station type':'StationType','Subscription status':'Subscription'})
renamed_df.head()
```

```python
# Convert the column header into a string 
renamed_df.columns = renamed_df.columns.astype("str") 

# Standardize the column header by Capitalizing the first letter of the column names
renamed_df.columns = renamed_df.columns.str.title()
renamed_df.head()
```

```python
# Combining day,month and year, hour and minute (in date time format - to_datetime)
# create a new column, "DateTime" with values from Year, Month, Day, Hour and Minute
# Using astype(str) to convert from int to String
renamed_df["DateTime"] = renamed_df['Year'].astype(str) + "-" + renamed_df['Month'].astype(str)+ "-" + renamed_df['Day'].astype(str) + "-" +renamed_df['Hour'].astype(str)+ "-" + renamed_df['Minute'].astype(str)
renamed_df.head()
```

**Research Question:** Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

```python
# Sort the day, hour and minute columns in ascending order
Time_df = renamed_df.sort_values(by = ["Day", "Hour","Minute"], ascending = True)
Time_df.head(10)
```

```python
# Filter the Time_df dataframe to show city as Paris
City_df =Time_df[Time_df.City == 'Paris']
City_df.head(10)
```

```python
# To differentiate the hours when the Bluecars were picked and returned
# We get the difference in blue cars using the diff() function
City_df["difference"] = City_df.Bluecars.diff()
City_df.head(15)
# For the output, when the difference is negative(-ve), it implies that the car has been picked. 
# If positive (+ve), the implication is that the car has been returned.
```

```python
# Finding the most popular hour Bluecars are picked.
Picking_df = City_df[(City_df['difference'] < 0)].groupby(["Hour"])["difference"].sum()
Picking_df.sort_values(ascending = False).head(1)
# We have the greatest difference at hour 18 meaning that,
# Most cars are picked from the station at 1800hrs
```

Therefore, the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018 is 1800hrs (06:00 pm)



*   What is the most popular hour for returning cars?





```python
# Getting the difference of Bluecars in the original data set - renamed_df
renamed_df["difference"] = renamed_df.Bluecars.diff()
# Most popular hour for returning Bluecars
returning_df = renamed_df[(renamed_df['difference'] > 0)].groupby(["Hour"])["difference"].sum()
returning_df.sort_values(ascending = False).head(1)
#  Most popular hours for returning Bluecars is 5:00am
```
